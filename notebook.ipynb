```python
import matplotlib.pyplot as plt
import numpy as np

# Zeitperioden
periods = ["Initiale Vorstellung", "1. WV", "2. WV", "3. WV"]

# Einnahmehäufigkeit in Prozent pro Periode
# Reihenfolge: Mehr als 3x täglich, 3x täglich, 2x täglich, 1x täglich, Mehrmals pro Woche, Keine Einnahme
data = {
    "Mehr als 3x täglich": [20, 18, 15, 12],
    "3x täglich": [50, 55, 60, 58],
    "2x täglich": [10, 8, 10, 12],
    "1x täglich": [8, 10, 8, 10],
    "Mehrmals pro Woche": [5, 4, 5, 6],
    "Keine Einnahme": [7, 5, 2, 2]
}

# Farben für die Einnahmehäufigkeit
colors = ['#FFC300', '#FF5733', '#C70039', '#900C3F', '#581845', '#1C1C1C']

# Balkendiagramm als gestapeltes Balkendiagramm
fig, ax = plt.subplots(figsize=(10, 6))

# Stapeln der Balken pro Einnahmehäufigkeit
bottom = np.zeros(len(periods))
for i, (key, values) in enumerate(data.items()):
    ax.bar(periods, values, bottom=bottom, label=key, color=colors[i])
    bottom += np.array(values)

# Beschriftung und Legende
ax.set_ylabel("Häufigkeit [%]")
ax.set_title("Abb3.: Einnahmehäufigkeiten der ASM der Erwachsenen mit PKU im Behandlungsverlauf")
ax.legend(title="Einnahmehäufigkeit", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig("Abb3a.png")
plt.show()
```

```python
# Erstellen eines Balkendiagramms, das geclusterte Balken für jede Einnahmehäufigkeit nach Zeit darstellt
fig, ax = plt.subplots(figsize=(10, 6))

# Anzahl der Einnahmekategorien
categories = list(data.keys())
num_categories = len(categories)
bar_width = 0.15  # Breite der einzelnen Balken
x = np.arange(len(periods))  # Positionen für die Zeitperioden auf der x-Achse

# Zeichnen der Balken für jede Einnahmehäufigkeit pro Zeitperiode
for i, (key, values) in enumerate(data.items()):
    ax.bar(x + i * bar_width, values, bar_width, label=key, color=colors[i])

# Anpassung der x-Achse
ax.set_xticks(x + bar_width * (num_categories - 1) / 2)
ax.set_xticklabels(periods)
ax.set_ylabel("Häufigkeit [%]")
ax.set_title("Abb3.: Einnahmehäufigkeiten der ASM der Erwachsenen mit PKU im Behandlungsverlauf")
ax.legend(title="Einnahmehäufigkeit", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig("Abb3b.png")
plt.show()
```

```python
# Diätverteilung in Prozent pro Periode
# Reihenfolge: Strikt berechnete, Geschätzte, Leicht reduzierte, Keine Diät
diet_data = {
    "Strikt berechnete Diät": [20, 31, 32, 33],
    "Geschätzte eiweißarme Diät": [36, 28, 27, 33],
    "Leicht reduzierte Eiweißzufuhr": [22, 38, 27, 33],
    "Keine Diät": [22, 3, 14, 0]
}

# Farben für die Diätverteilung (Gelb- und Rottöne)
diet_colors = ['#FFC300', '#FF5733', '#C70039', '#900C3F', '#581845', '#1C1C1C']

# Balkendiagramm als gestapeltes Balkendiagramm für Diätverteilung
fig, ax = plt.subplots(figsize=(10, 6))

# Stapeln der Balken pro Diätkategorie
bottom = np.zeros(len(periods))
for i, (key, values) in enumerate(diet_data.items()):
    ax.bar(periods, values, bottom=bottom, label=key, color=diet_colors[i])
    bottom += np.array(values)

# Beschriftung und Legende
ax.set_ylabel("Prozentuale Verteilung [%]")
ax.set_title("Abb2: Durchgeführte eiweißarme bzw. -reduzierte Diäten der Erwachsenen mit PKU im Behandlungsverlauf (2019-2024).")
ax.legend(title="Diätart", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig("Abb2a.png")
plt.show()
```

```python
# Erstellung eines geclusterten Balkendiagramms für die Diätverteilung pro Zeitperiode
fig, ax = plt.subplots(figsize=(10, 6))

# Anzahl der Diätkategorien
diet_categories = list(diet_data.keys())
num_diet_categories = len(diet_categories)
x = np.arange(len(periods))  # Positionen für die Zeitperioden auf der x-Achse

# Zeichnen der Balken für jede Diätkategorie pro Zeitperiode
for i, (key, values) in enumerate(diet_data.items()):
    ax.bar(x + i * bar_width, values, bar_width, label=key, color=colors[i])

# Anpassung der x-Achse
ax.set_xticks(x + bar_width * (num_diet_categories - 1) / 2)
ax.set_xticklabels(periods)
ax.set_ylabel("Prozentuale Verteilung [%]")
ax.set_title("Diätverteilung in Prozent geclustert nach Zeitperioden")
ax.grid(True, which='major', axis='y', linestyle='--', alpha=0.7, zorder=0)
ax.legend(title="Diätart", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig("Abb2b.png")
plt.show()
```
